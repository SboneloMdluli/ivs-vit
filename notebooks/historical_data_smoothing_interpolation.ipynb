# Full historic surface pipeline → unified grid → normalization → PDF report

Everything above shows individual kernel-smoothing experiments. The cells below run the **full pipeline** that every downstream consumer should use for historic data:

1. **Smoothing** — Gaussian-kernel weighted average of raw quotes.
2. **Interpolation** — linear 2D interpolation of NaN cells inside the convex hull of quotes.
3. **Extrapolation** — nearest-neighbour fill for any remaining NaN cells in the wings.
4. **Unified grid** — resample onto the canonical `(k, τ)` axes from `config/unified_iv_grid.yaml`, so pixel `(i, j)` always maps to the same `(k_i, τ_j)` cell as every synthetic SABR / Heston surface.
5. **Normalization** — per-pixel `log σ` z-score via the pre-fit `SurfaceNormalizer` shipped at `data/processed/iv_normalizer.npz`.

The final cell writes a focused **2-page PDF**: page 1 stacks the 3D plots of the raw quotes, the kernel-smoothed surface, and the normalized surface; page 2 shows the same three artefacts as heatmaps.

> **Note — "why do raw quotes look the same across dates?"**  
> The raw scatter cloud always fills roughly the same `(k, τ)` envelope because `filter_day_for_surface` clips every day to the same `k ∈ [-0.5, 0.5]` / `τ ∈ [0.01, 2.0]` box, and each trading day already contains 10k+ quotes that saturate that box. The IV values (z-axis / colour) _do_ change day-to-day; with the PDF now using a shared IV colour scale across the raw scatter and the smoothed surface, the day-to-day level/skew shifts become visible. If the PDF looks unchanged after editing `sample_date` below, make sure to re-run the pipeline cell (not only the save cell), since `stages` is recomputed there.

```python
from implied_volatility_diffusion import SurfaceNormalizer, UnifiedGrid
from implied_volatility_diffusion.data.historical_data_smoothing_interpolation import (
    build_historical_surface_pipeline,
)

unified_grid = UnifiedGrid.load(PROJECT_ROOT / "config" / "unified_iv_grid.yaml")
print("unified grid shape:", unified_grid.shape)

normalizer_path = PROJECT_ROOT / "data" / "processed" / "iv_normalizer.npz"
normalizer = SurfaceNormalizer.load(normalizer_path) if normalizer_path.exists() else None
print("normalizer:", "loaded" if normalizer is not None else "not available (skip normalization)")
```

```python
stages = build_historical_surface_pipeline(
    day,
    sample_date=sample_date,
    unified_grid=unified_grid,
    x_grid_smooth=np.linspace(unified_grid.log_moneyness[0], unified_grid.log_moneyness[-1], 61),
    tau_grid_smooth=np.linspace(unified_grid.tau[0], unified_grid.tau[-1], 60),
    weight_col="vega",
    clip_upper=50.0,
    h1=0.01,
    h2=0.10,
    normalizer=normalizer,
)

nan_before = int(np.count_nonzero(~np.isfinite(stages.smoothed)))
nan_after_fill = int(np.count_nonzero(~np.isfinite(stages.filled)))
nan_after_unified = int(np.count_nonzero(~np.isfinite(stages.iv_unified)))
print(f"smoothed NaNs              : {nan_before} / {stages.smoothed.size}")
print(f"after interp/extrap NaNs   : {nan_after_fill} / {stages.filled.size}")
print(f"unified-grid NaNs          : {nan_after_unified} / {stages.iv_unified.size}")
print("unified surface shape      :", stages.iv_unified.shape)
if stages.iv_normalized is not None:
    print(
        "normalized z-score range   :",
        float(np.nanmin(stages.iv_normalized)),
        "..",
        float(np.nanmax(stages.iv_normalized)),
    )
```

```python
from implied_volatility_diffusion.data.historical_data_smoothing_interpolation import (
    make_surface_comparison_3d_figure,
    make_surface_comparison_heatmap_figure,
)

if stages.iv_normalized is None:
    raise RuntimeError(
        "stages.iv_normalized is None — a SurfaceNormalizer must be supplied to "
        "build_historical_surface_pipeline for the normalized z-score surface. "
        "Re-run the cell above with a loaded normalizer."
    )

iv_raw_day = stages.day_sub["iv"].to_numpy(dtype=float)
print(f"sample_date                  : {stages.sample_date.date()}")
print(f"raw quotes used              : {len(stages.day_sub)}")
print(
    f"raw iv  (mean / p99 / max)   : "
    f"{float(np.nanmean(iv_raw_day)):.4f} / "
    f"{float(np.nanpercentile(iv_raw_day, 99.0)):.4f} / "
    f"{float(np.nanmax(iv_raw_day)):.4f}"
)
print(
    f"iv_unified    (min .. max)   : "
    f"{float(np.nanmin(stages.iv_unified)):.4f} .. "
    f"{float(np.nanmax(stages.iv_unified)):.4f}"
)
print(
    f"iv_normalized (min .. max)   : "
    f"{float(np.nanmin(stages.iv_normalized)):.3f} .. "
    f"{float(np.nanmax(stages.iv_normalized)):.3f}"
)

fig_3d = make_surface_comparison_3d_figure(stages)
plt.show()

fig_heat = make_surface_comparison_heatmap_figure(stages)
plt.show()
```

```python
from implied_volatility_diffusion.data.historical_data_smoothing_interpolation import (
    save_surface_comparison_pdf,
)

reports_dir = PROJECT_ROOT / "data" / "processed" / "reports"
pdf_path = reports_dir / f"historic_surface_comparison_{sample_date.date()}.pdf"

save_surface_comparison_pdf(stages, pdf_path)
print("wrote report to:", pdf_path)
```
